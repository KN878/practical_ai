```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original = cv2.cvtColor(cv2.imread('1957.png', cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB)
img = cv2.imread('1957.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap="gray")
plt.show()
```

# Binarization of image

```python
ret,binary = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(binary, cmap='gray')
plt.show()
```

# Dilation

```python
kernel = np.ones((4,4),np.uint8)
#removing holes in digits
dilated = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)
#removing noise on the backgroung
dilated = cv2.morphologyEx(dilated, cv2.MORPH_OPEN, kernel)
#making digits thicker
kernel = np.ones((6,6),np.uint8)
dilated = cv2.erode(dilated, kernel, iterations = 1)
plt.imshow(dilated, cmap='gray')
plt.show()
```

# Drawing bounding boxes and extracting digits

```python
contours, hier = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
digits = []
for c in contours:
    x,y,w,h = cv2.boundingRect(c) 
    borderedCopy = cv2.copyMakeBorder(dilated[y:y+h, x:x+w], 20, 20, 20, 20, cv2.BORDER_CONSTANT)
    digits.append(borderedCopy)
    original = cv2.rectangle(original,(x,y),(x+w,y+h),(0,255,0),2)
    
    
plt.imshow(original, cmap = 'gray')
plt.show()
for digit in digits:
    plt.imshow(digit, cmap='gray')
    plt.show()
```

# Training classifier

```python
from keras.datasets import mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()
# save input image dimensions
img_rows, img_cols = 28, 28

X_train = X_train / 255.
X_test = X_test / 255.

X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
```

```python
from keras.utils import to_categorical
num_classes = 10

y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D

model = Sequential()
model.add(Conv2D(32, (3, 3), activation = 'relu', input_shape = (28, 28, 1)))
model.add(MaxPooling2D(2, 2))
model.add(Dropout(0.5))
model.add(Conv2D(64, (3, 3), activation = 'relu'))
model.add(MaxPooling2D(2, 2))
model.add(Dropout(0.5))
model.add(Conv2D(64, (3, 3), activation = 'relu'))
model.add(Flatten())
model.add(Dense(128, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation = 'softmax'))
model.compile(optimizer = 'rmsprop', loss = 'categorical_crossentropy', metrics = ['acc'])
```

```python
model.fit(X_train, y_train, epochs = 20, batch_size = 128, validation_data = (X_test, y_test), verbose = 2)
```

# Predicting digits

```python
predictions = np.array([])
for digit in digits:
    resized = cv2.resize(digit, (28, 28), interpolation = cv2.INTER_AREA)
    pred = model.predict_classes(resized.reshape(-1, 28, 28, 1))
    predictions = np.append(predictions, pred)

sorted_pred = np.sort(predictions.flatten())
print("Predicted digits: {}".format(sorted_pred))
```
