```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture("cut.mp4")
frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
fps = cap.get(cv2.CAP_PROP_FPS)
duration = float(frame_count) / float(fps)
video = []

ret = True
while(ret):
  ret, frame = cap.read()
  if ret:
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    video.append(frame)
H, W, _ = video[0].shape
cap.release()
```

```python
print(f"{W}x{H}, {duration:.2f} seconds, {fps:.3f} frames per second")
```

```python
# the pretrained Xception model takes 299x299, whichc is square, while our video is not
# to solve this, I perform sliding window, mapping each frame to several square frames

def sliding_window(frame):
  result = []
  if W > H:
    step = H // 2
    x = 0
    while x + H <= W:
      result.append(frame[:, x:x + H])
      x += step
    result.append(frame[:, W - H:])
  elif W < H:
    step = W // 2
    x = 0
    while x + W <= H:
      result.append(frame[x:x + W, :])
      x += step
    result.append(frame[H - W:, :])
  else:
    result.append(frame)
  return result


sw_video = []
for f in video:
  sw_video.extend(sliding_window(f))

sw_video = [cv2.resize(f, (224, 224)) for f in sw_video]
video_processed = preprocess_input(np.array(sw_video))
```

```python
# GPU recommended
from tensorflow.keras.applications.vgg16 import decode_predictions, preprocess_input, VGG16

model = VGG16()
# predict the probability across all output classes
yhat = model.predict(video_processed)
# convert the probabilities to class labels
sw_prediction = decode_predictions(yhat)
```

```python
from collections import defaultdict

# reduce back
k = len(sw_video) // len(video)
prediction = [[] for _ in video]
for i, f in enumerate(sw_prediction):
  prediction[i // k].extend(f)

def reduce(pred):
  conf = defaultdict(float)
  num = defaultdict(int)
  for x in pred:
    conf[(x[0], x[1])] += x[2]
    num[x[1]] += 1
  return [(k1, k2, v / num[k2]) for (k1, k2), v in conf.items()]

# sort labels by confidence
prediction = [sorted(reduce(f), key=lambda x: x[2], reverse=True) for f in prediction]
```

```python
exceptions = ["web_site", "television", "home_theater", "monitor", "balance_beam", "tench"]
dog = {"Labrador_retriever", "kuvasz", "Great_Pyrenees", "dingo",
       "borzoi", "white_wolf", "patas", "Siberian_husky", "Bedlington_terrier"}
bone = {"walking_stick", "crutch", "ping-pong_ball"}
player = {"ballplayer"}



def get_label(label_pred):
  # get most probable label which is not in the exceptions list
  label = label_pred[0]
  j = 1
  while j < len(label_pred) and label[1] in exceptions:
    label = label_pred[j]
    j += 1

  if j == len(label_pred):
    return None
  
  if label[1] in dog:
    return (label[1], "dog", label[2])
  if label[1] in bone:
    return (label[1], "bone", label[2])
  if label[2] > 0.05:
    return label


def print_interval(i, l, last_time):
  def format_time(x):
    return f"{x / 24:.1f}"
  print(f"{format_time(last_i)}-{format_time(i)}: Label {l[1]}, \tConfidence {l[2]:.3f}")


labels = []
# print intervals with labels
last, last_i = None, 0
for i in range(len(video)):
  label = get_label(prediction[i])
  labels.append(label)

  if not label:
    continue

  if not last:
    last = label[1]

  if last and last != label[1]:
    print_interval(i, label, last_i)
    last, last_i = label[1], i

if label and last and last != label[1]:
  print_interval(i, label[1], last_i)
```
