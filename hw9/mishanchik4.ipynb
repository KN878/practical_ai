```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
img = (cv2.cvtColor(cv2.imread('graph.png'), cv2.COLOR_BGR2GRAY) < 240).astype('uint8') * 255
kernel = np.ones((2, 2), np.uint8)
img = cv2.erode(img, kernel, iterations=5)
img = cv2.dilate(img, kernel, iterations=6)
```

```python
edge_kernel = np.ones((2,2), np.uint8)
```

```python
plt.imshow(img, cmap="gray")
plt.show()
edges = cv2.Canny(img, 50, 50)
edges = cv2.dilate(edges, edge_kernel, iterations=5)
cimg = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
plt.imshow(edges)
plt.show()

# TODO: restore the nodes and the edges or the graph by picture
```

```python
circles = cv2.HoughCircles(edges, 
                   cv2.HOUGH_GRADIENT,
                   dp=1,       # internal ratio parameter, usually 1
                   minDist=20, # between centers
                   param1=254,  # parameters of mode (canny edge)
                   param2=17,  # bigger - less false positives
                   minRadius=10,
                   maxRadius=60)
```

```python
al = np.around(circles)
circles = np.uint16(al)
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(cimg, (*i[:2],), i[2], (255,0,0), 2)
    # draw the center of the circle
    cv2.circle(cimg, (*i[:2],), 2, (0,0,255), 3)
plt.imshow(cimg)
plt.show()
```

```python
img = (cv2.cvtColor(cv2.imread('graph.png'), cv2.COLOR_BGR2GRAY) < 240).astype('uint8') * 255
plt.imshow(img)
```

```python
kernel = np.ones((2,2), np.uint8)
```

```python
test = cv2.dilate(img, kernel, iterations=2)
```

```python
low_threshold = 50
high_threshold = 150
edges = cv2.Canny(test, low_threshold, high_threshold)
test = edges
```

```python
plt.imshow(test)
```

```python
rho = 1
theta = np.pi / 180 
threshold = 30 
min_line_length = 50 
max_line_gap = 20 
line_image = np.copy(test) * 0  

lines = cv2.HoughLinesP(test, rho, theta, threshold, np.array([]),
                    min_line_length, max_line_gap)

for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),5)
```

```python
plt.imshow(line_image)
```

```python
def create_graph(circles, lines):
    
    def distance(x1,y1,x2,y2):
        
        return  np.sqrt((x1-x2)**2 + (y1-y2)**2)
    
    circles = circles[0]
    R = circles[0][2] # ~same for all circles
    threshold = np.sqrt((R+10)**2 + (R+10)**2)
    G = nx.Graph()
    
    for x, y, r in circles:
        
        G.add_node((x,y))
        
    for line in lines:
        
        xStart, yStart, xEnd, yEnd = line[0]
        startNode, endNode = None, None
        
        for x, y in G.nodes():
            
            if distance(x, y, xStart, yStart) < threshold:
                
                startNode = (x,y)
                
            if distance(x, y, xEnd, yEnd) < threshold:
                
                endNode = (x,y)
                
        if (startNode != None and endNode != None):
            
            G.add_edge(startNode,endNode)
    
    return G
```

```python
GG = create_graph(circles, lines)
```

```python
nx.draw(GG)
```
