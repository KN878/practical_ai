```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('fish.jpg', cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
img_smooth = cv2.medianBlur(img_rgb, 7) # or bilateral, it is slower
plt.imshow(img_smooth, cmap='gray')
plt.show()
```

```python
lower_yellow = np.array([105, 0, 0])
upper_yellow = np.array([255, 255, 255])
mask = cv2.inRange(img_smooth, lower_yellow, upper_yellow)
res = cv2.bitwise_and(img_smooth,img_smooth, mask= mask)
plt.imshow(res)
plt.show()
```

```python
img_gray = cv2.cvtColor(res, cv2.COLOR_RGB2GRAY)
threshold = [20, 255]
output = np.zeros_like(img_gray)
output[(img_gray >= threshold[0]) & (img_gray <= threshold[1])]=1
plt.imshow(output, cmap='gray')
plt.show()
```
